```python
import geemap.geemap as geemap
import ee
import os
```

```python
ee.Authenticate()
ee.Initialize(project = 'neon-feat-313910')
```

```python
base = geemap.Map()
Sangola_aoi = ee.FeatureCollection('projects/neon-feat-313910/assets/Sangola')
base.addLayer(Sangola_aoi, {}, 'Sangola_aoi')
# base
```

```python
Sangola_NDVI_2002 = ee.ImageCollection('LANDSAT/COMPOSITES/C02/T1_L2_8DAY_NDVI').filterDate('2002-04-01', '2002-12-31').mean().clip(Sangola_aoi).select('NDVI')
Mulshi_vis = {
  'min': 0.1,
  'max': 0.5,
  'palette': ['ffffff', 'ce7e45', 'df923d', 'f1b555', 'fcd163', '99b718', '74a901',
    '66a000', '529400', '3e8601', '207401', '056201', '004c00', '023b01',
    '012e01', '011d01', '011301']}
base.addLayer(Sangola_NDVI_2002, Mulshi_vis, 'Sangola 2002')
# base
```

```python
Sangola_NDVI_2010 = ee.ImageCollection('LANDSAT/COMPOSITES/C02/T1_L2_8DAY_NDVI').filterDate('2010-01-01', '2010-12-31').mean().clip(Sangola_aoi).select('NDVI')
Mulshi_vis = {
  'min': 0.1,
  'max': 0.4,
  'palette': ['ffffff', 'ce7e45', 'df923d', 'f1b555', 'fcd163', '99b718', '74a901',
    '66a000', '529400', '3e8601', '207401', '056201', '004c00', '023b01',
    '012e01', '011d01', '011301']}
base.addLayer(Sangola_NDVI_2010, Mulshi_vis, 'Sangola 2010')
# base
```

```python
Sangola_NDVI_2024 = ee.ImageCollection('LANDSAT/COMPOSITES/C02/T1_L2_8DAY_NDVI').filterDate('2024-03-01', '2024-12-31').mean().clip(Sangola_aoi).select('NDVI')
Mulshi_vis = {
  'min': 0.0,
  'max': 0.5,
  'palette': ['ffffff', 'ce7e45', 'df923d', 'f1b555', 'fcd163', '99b718', '74a901',
    '66a000', '529400', '3e8601', '207401', '056201', '004c00', '023b01',
    '012e01', '011d01', '011301']}
base.addLayer(Sangola_NDVI_2024, Mulshi_vis, 'Sangola 2024')
base.centerObject(Sangola_aoi, 10)
base
```
